```python
import pandas as pd
import pymongo as pymongo
from datetime import datetime
from datetime import timedelta
```

```python
# Set up MongoDB
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Declare the database
db = client.speedskate_db

# Declare the collection
collection = db.speedskate_db
```

```python
csv_file = "ladies-1500-2019-2020.csv"
speed_df = pd.read_csv(csv_file)
speed_df.head()
```

```python
# Add column with distance
speed_df['distance'] = '1500'
speed_df
```

```python
# get Time into Proper Format
def clean(x):
    '''function to clean results csv'''
    #Convert time to proper form and cast to float
    x['time']= x['time'].str.replace('.', ':')
    x['time']= x['time'].str.replace(',', ':')
    x['time']= x['time'].str.replace(' ', '')
    # Drop column
    #x=x.drop(columns = ['Unnamed: 5'])
    # Rename columns
    #x=x.rename(columns={'Unnamed: 2':'Age'})
    return x.head()

clean(speed_df)
```
